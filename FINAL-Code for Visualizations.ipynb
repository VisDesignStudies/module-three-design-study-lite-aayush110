```python
import pandas as pd
df = pd.read_csv('energy.csv')
# df = data.reset_index()
df.head()
```

```python
import altair as alt

# Visualization A
# Assuming df is your DataFrame
df_reset_index = df.reset_index()


click = alt.selection_multi(encodings=['color'])

scatter = alt.Chart(df_reset_index).mark_point().encode(
    x=alt.X('level_8:Q', title="Consumption.Commercial.Petroleum"),
    y=alt.Y('level_22:Q', title="Consumption.Industrial.Petroleum"),
    color=alt.Color('level_1:N', title='Year'),
    tooltip=[alt.Tooltip('level_0:N', title='State')]
).transform_filter(
    click
).add_selection(
    click
)

hist = alt.Chart(df_reset_index).mark_bar().encode(
    x=alt.X('level_1:Q', title="Year", axis=alt.Axis(format=".0f", tickMinStep=1, tickCount=len(df_reset_index))),
    y=alt.Y('level_22:Q', title="Consumption.Industrial.Petroleum"),
    color=alt.Color('level_1:N', title='Year'),
    tooltip=[alt.Tooltip('level_1:O', title='Year')]
).add_selection(
    click
)

(scatter & hist)
```

```python
import altair as alt

# Visualization B
# Assuming df is your DataFrame
df_reset_index = df.reset_index()


click = alt.selection_multi(encodings=['color'])

scatter_b = alt.Chart(df_reset_index).mark_point().encode(
    x=alt.X('level_3:Q', title="Consumption.Commercial.Coal"),
    y=alt.Y('level_17:Q', title="Consumption.Industrial.Coal"),
    color=alt.Color('level_1:N', title='Year'),
    tooltip=[alt.Tooltip('level_0:N', title='State')]
).transform_filter(
    click
).add_selection(
    click
)

hist_b = alt.Chart(df_reset_index).mark_bar().encode(
    x=alt.X('level_1:Q', title="Year", axis=alt.Axis(format=".0f", tickMinStep=1, tickCount=len(df_reset_index))),
    y=alt.Y('level_8:Q', title="Consumption.Commercial.Coal"),
    color=alt.Color('level_1:N', title='Year'),
    tooltip=[alt.Tooltip('level_1:O', title='Year')]
).add_selection(
    click
)

(scatter_b & hist_b)
```

```python
#COMMERCIAL PETROLEUM VS SOLAR CONSUMPTION BY STATE - VISUALIZATION C IN OUR PROTOTYPE

# Choose the State
choose_state = alt.selection_single(
    name='Choose',
    fields=['State'],
    bind=alt.binding_select(options=df['State'].unique().tolist()),
    init={'State': df['State'].unique().tolist()[0]}
)

# Display chart
chart1 = alt.Chart(df).mark_line().encode(
    x=alt.X('Year', title='Year', axis=alt.Axis(format='d')),  
    y=alt.Y('ConsumptionCommercialPetroleum', title='Energy Consumption (Billion Btu)'),
    color=alt.value('red'),  # use a static color value
    tooltip=['Year', 'ConsumptionCommercialPetroleum', 'State']
)

chart2 = alt.Chart(df).mark_line().encode(
    x=alt.X('Year', title='Year', axis=alt.Axis(format='d')),  
    y=alt.Y('ConsumptionCommercialSolar'),
    color=alt.value('blue'),  # use a static color value
    tooltip=['Year', 'ConsumptionCommercialSolar', 'State']
)

overlayed_chart = (chart1 + chart2).add_selection(choose_state).transform_filter(choose_state).properties(
    title='Commercial Petroleum vs Solar Consumption by State since 1960',
    width=600,
    height=400
).interactive()

# Add axis labels and legend titles
final_chart = overlayed_chart.configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_legend(
    labelFontSize=12,
    titleFontSize=14,
   orient='bottom'  # change legend orientation to horizontal
)

final_chart
```
